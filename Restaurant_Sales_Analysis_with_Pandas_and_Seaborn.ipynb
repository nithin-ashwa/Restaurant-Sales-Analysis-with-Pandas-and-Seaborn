```python
import numpy as np
import pandas as pd
```

```python
c = data.groupby('ITEM')['SUB_TOTAL'].sum().sort_values(ascending=False).head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(c)
```
